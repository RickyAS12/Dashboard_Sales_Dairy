```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dd =pd.read_csv('dairy_dataset.csv')
```

```python
dd.head()
```

```python
dd.shape
```

```python
# Looks like no null or NaN data
dd.info()
```

```python
numerik_column = []
kategorik_column = []
```

```python
for column in dd.columns.tolist():
  if dd[column].dtypes == 'float64' or dd[column].dtypes == 'int64':
    numerik_column.append(column)
  else:
    kategorik_column.append(column)
```

```python
dd[numerik_column].describe()
```

```python
dd[kategorik_column].describe()
```

```python
# Few columns have outliers
plt.figure(figsize=(12, 6))
for i in range(len(numerik_column)):
  plt.subplot(2, len(numerik_column)//2 + 1, i+1)
  sns.boxplot(data=dd, x=dd[numerik_column[i]])
  sns.stripplot(x=dd[numerik_column[i]], data=dd, color='red', jitter=True, size=8)
  plt.tight_layout()
```

```python
plt.figure(figsize=(12, 6))
for i in range(len(numerik_column)):
  plt.subplot(2, len(numerik_column)//2 + 1, i+1)
  sns.kdeplot(data=dd, x=dd[numerik_column[i]], fill=True)
  plt.tight_layout()
```

```python
features = kategorik_column
plt.figure(figsize=(12, 6))
for i in range(0, len(features)):
    plt.subplot(1, len(features), i+1)
    sns.countplot(x=dd[features[i]])
    plt.xlabel(features[i])
    plt.tight_layout()
```
